```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
music_csv = Path("multiTimeline.csv")

music_data = pd.read_csv(music_csv)

print(f"Number of months: {len(music_data)}")
music_data.head()
```

```python
# too many months. group by year
music_data = music_data.rename(columns={"music: (United States)":"Interest"})
music_data["Year"] = music_data["Month"].str.slice(0,4)
music_data.head()
```

```python
# create by year dataset
music_by_year = music_data.groupby(["Year"]).mean(numeric_only=True).round(2)
print(f"Number of by years: {len(music_by_year)}")
music_by_year.head()

```

```python
music_bar_chart = music_by_year.plot(kind="bar")

music_bar_chart.set_xlabel("Timeline")
music_bar_chart.set_ylabel("Interest in USA")

plt.show()

plt.close()
```

```python
music_pie_chart = music_by_year.plot(kind='pie', subplots=True)

fig = plt.gcf()
fig.set_size_inches(9,5)
# fig = plt.figure(figsize=(55,30))


plt.axis("equal")

# music_pie_chart.set_ylabel("Month")

plt.show()
plt.close()
```

```python

music_year = music_data.groupby("Year")[["Interest"]].mean()
music_intestest = music_data["Interest"].mean()

plt.scatter(music_year, music_intestest)

plt.xlabel("Popularity")
plt.ylabel("Time")

plt.show()

plt.close()
```
