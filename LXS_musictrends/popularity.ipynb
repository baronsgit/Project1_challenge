```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
song_csv = Path("song_data.csv")

song_data = pd.read_csv(song_csv)

print(f"Number of songs: {len(song_data)}")

song_data.head()
```

```python
colName1 = "song_popularity"
colName2 = "danceability"
```

```python

song_grp = song_data.groupby("key")[[colName1, colName2]].mean()

duration_ave = song_grp[colName1]
pop_ave = song_grp[colName2]

plt.scatter(duration_ave, pop_ave)

plt.xlabel(colName1)
plt.ylabel(colName2)

plt.show()
plt.close()
```

```python
song_bar_chart = song_grp.plot(kind="bar")

song_bar_chart.set_xlabel(colName1)
song_bar_chart.set_ylabel(colName2)

plt.show()

plt.close()
```
